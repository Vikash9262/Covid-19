```python
sns.lineplot(x='New deaths',y='New recovered' , data = df)
plt.title("New Deaths vs New Recovered" , fontsize = 20)
plt.show()
```

```python
sns.lineplot(x='New deaths',y='New recovered' , data = df , hue='WHO Region')
plt.title("New Deaths vs New Recovered" , fontsize = 20)
plt.show()
```

```python
sns.lineplot(x='New deaths',y='New recovered' , data = df , hue='WHO Region', style='WHO Region')
plt.title("New Deaths vs New Recovered" , fontsize = 20)
plt.show()
```

```python
sns.lineplot(x='New deaths',y='New recovered' , data = df , hue='WHO Region', style='WHO Region',markers=True)
```

```python
Sales = df.groupby("WHO Region")["New recovered"].sum().sort_index()
plt.figure(figsize = (9,5))
bars = plt.plot(Sales.index , Sales.values , marker='o',linestyle="--",color='red',linewidth=2)

plt.xlabel("WHO Region",fontsize=15)
plt.ylabel("New recovered",fontsize=15)
plt.title("WHO Region by New recovered",fontsize=23)

plt.tight_layout()
plt.show()
```

# Bar Plot

```python
plt.figure(figsize = (15,6))
sns.barplot(x=df.Deaths  , y=df.Recovered)
plt.title("Deaths and Recovered Bar Graph",fontsize = 30)
plt.xlabel("Deaths",fontsize = 30)
plt.ylabel("Recovered",fontsize = 30)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
Bar = df['WHO Region'].value_counts()
Bar.plot(kind='bar',color=['red','green','yellow','blue','pink','black'])
plt.title("WHO Region bar Graph",fontsize = 23)
plt.xlabel("WHO Region",fontsize = 15)
plt.ylabel("Counts",fontsize = 15)
plt.tight_layout()
plt.xticks(rotation=-40)
plt.show()
```

```python
Bar = df['WHO Region'].value_counts()
Bar.plot(kind='pie',autopct = '%1.1f%%' , color=['red','green','yellow','blue','pink','black'])
plt.title("WHO Region bar Graph",fontsize = 23)
plt.tight_layout()

plt.show()
```

# WHO Pie Chart Based on New RECOVERED

```python
plt.figure(figsize=(5,5))
Cases= df.groupby("WHO Region")['New recovered'].sum()
plt.pie(Cases,labels=Cases.index , autopct='%1.1f%%')
plt.title("WHO Pie Chart ")
plt.show()
```

# WHO Pie Chart Based on New Cases

```python
plt.figure(figsize=(5,5))
Cases= df.groupby("WHO Region")['New cases'].sum()
plt.pie(Cases,labels=Cases.index , autopct='%1.1f%%')
plt.title("WHO Pie Chart ")
plt.show()
```

# WHO Pie Chart Based on New Deaths

```python
plt.figure(figsize=(5,5))
Cases= df.groupby("WHO Region")['New deaths'].sum()
plt.pie(Cases,labels=Cases.index , autopct='%1.1f%%')
plt.title("WHO Pie Chart ")
plt.show()
```

```python
plt.figure(figsize=(5,5))
Cases= df.groupby("WHO Region")['New deaths'].sum()
myexplode = [0,0,0.2,0,0,0]
plt.pie(Cases,labels=Cases.index , autopct='%1.1f%%',explode=myexplode)
plt.title("WHO Pie Chart ")
plt.show()
```

```python
Grouped = df.groupby(["New deaths","WHO Region"])["New recovered"].sum().unstack()

ax= Grouped.plot(kind='bar',figsize = (6,6) , title = "Outlet Tier by Item Fat Content")
plt.xlabel("Outlet Location Type")
plt.ylabel("Item Fat Content ")
plt.legend(title = "Item Fat Content")
plt.tight_layout()
plt.show()
```
